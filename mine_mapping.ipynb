# Создание цифровых карт шахты с помощью LiDAR и фотограмметрии

```python
import numpy as np
import matplotlib.pyplot as plt
print('Проект: Создание цифровых карт шахты')
print('Метод: LiDAR и фотограмметрия')
```

```python
np.random.seed(42)
x = np.random.normal(0, 10, 3000)
y = np.random.normal(0, 5, 3000)
z = np.random.normal(-20, 3, 3000)
print('Количество точек:', len(x))
```

```python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z, s=1)
ax.set_title('Облако точек шахты')
plt.show()
```

```python
theta = np.linspace(0, 2*np.pi, 100)
z = np.linspace(-50, 0, 100)
theta, z = np.meshgrid(theta, z)
r = 5
x = r*np.cos(theta)
y = r*np.sin(theta)
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x,y,z)
ax.set_title('Модель шахтного ствола')
plt.show()
```

```python
print('Вывод: смоделирован процесс построения цифровой модели шахты.')
```
